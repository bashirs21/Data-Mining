<h1>Tokenization</h1>


Tokenization splits a stream of text into meaningful tokens. A tokenizer divides a string into substrings by splitting on the specified string. 

Tokenizer is used here to split the data into meaningful tokens and remove punctuations. Each token is separated by comma (,) :

<h1> Frequency Distribution </h1>


<b>NLTK Frequency distribution function</b> is used to record frequency of each word in a document, given its length.


```python
import nltk

from nltk.tokenize import word_tokenize, RegexpTokenizer
from nltk.corpus import stopwords
tokenizer = RegexpTokenizer(r'\w+')
```

```python
 
def process_text(tweets):
    tokens = []
    for col in tweets:
        toks = tokenizer.tokenize(col)
        tokens.extend(toks)
    return tokens

col = list(df['data_without_stopwords'])
tokens = process_text(col)
freq = nltk.FreqDist(tokens)
```

```python
freq_words = freq.most_common(50)
freq_words
```

This is awesome! we revealed the most frequently used words in the tweets. The word 'coronavirus' and 'Trump' were the most popular words.


Now, its time to visualize the results. The below code will display the bar graph with the most frequently used 50 words in the tweets

```python
from matplotlib import pyplot as plt


x_labels = [val[0] for val in freq_words]
y_labels = [val[1] for val in freq_words]
plt.figure(figsize=(12, 6))
ax = pd.Series(y_labels).plot(kind='bar')
ax.set_xticklabels(x_labels)
ax.set_title('The most 50 frequent words')
rects = ax.patches

for rect, label in zip(rects, y_labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height, label, ha='center', va='bottom')
```

<h2> Challenge Question 5 </h2>

Visualize only the top 10 frequently used words 

```python
freq_words2 = freq.most_common(10)
freq_words2       

from matplotlib import pyplot as plt


x_labels = [val[0] for val in freq_words2]
y_labels = [val[1] for val in freq_words2]
plt.figure(figsize=(12, 7))
ax = pd.Series(y_labels).plot(kind='bar')
ax.set_xticklabels(x_labels)
ax.set_title('The 10 most frequent words')
rects = ax.patches

for rect, label in zip(rects, y_labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height, label, ha='center', va='bottom')
```
